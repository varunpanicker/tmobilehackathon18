```python
def myfillna(series):
    if series.dtype is pd.np.dtype(float):
        return series.fillna('')
    elif series.dtype is pd.np.dtype(int):
        return series.fillna('')
    else:
        return series.fillna('NA')
```

## Deep learning  for prediction

```python
from keras.preprocessing import sequence
from keras.models import Sequential
from keras.layers import Activation, Dense, Embedding, Bidirectional
from keras.layers import LSTM
from keras.datasets import imdb
from keras.layers import Flatten
from keras.layers import TimeDistributed
from keras.layers import AveragePooling1D
```

```python
from keras.models import load_model
```

```python
data.rename(columns={'text':'comment_text'},inplace=True)
```

```python
train_size = int(len(data) * .2)
train_posts = data['comment_text'][train_size:]
train_tags = data['rating'][train_size:]
test_posts = data['comment_text'][:train_size]
test_tags = data['rating'][:train_size]
```

```python
from keras.preprocessing.text import Tokenizer
```

```python
vocab_size = 2000
tokenize = Tokenizer(num_words=vocab_size)
tokenize.fit_on_texts(train_posts)
```

```python
x_train = tokenize.texts_to_matrix(train_posts)
```

```python
from sklearn.preprocessing import LabelBinarizer
```

```python
train_tags.shape
```

```python
train_tags=train_tags.astype(int)
test_tags=test_tags.astype(int)
```

```python
encoder = LabelBinarizer()
encoder.fit(train_tags)
y_train = encoder.transform(train_tags)
y_test = encoder.transform(test_tags)
```

```python
x_train = np.reshape(x_train, (x_train.shape[0], 1, x_train.shape[1]))
```

```python
x_train.shape
```

```python
from keras.utils import to_categorical
y_train = to_categorical(y_train)
```

## Recurrent neural networks using LSTM

```python
data_dim = 2000
timesteps = 1
num_classes = 2

model = Sequential()
model.add(Bidirectional(LSTM(30, return_sequences=True,
               input_shape=(timesteps, data_dim))))  # returns a sequence of vectors of dimension 30
model.add(Bidirectional(LSTM(30, return_sequences=True)))  # returns a sequence of vectors of dimension 30
model.add(Bidirectional(LSTM(30)))  # return a single vector of dimension 30
model.add(Activation('relu'))
model.add(Dense(2, activation='softmax'))

model.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])

```

```python
model.fit(x_train, y_train, batch_size = 100, epochs = 150, verbose = 1)
```

```python
model.summary()
```

```python
x_test=tokenize.texts_to_matrix(test_posts)
x_test = np.reshape(x_test, (x_test.shape[0], 1, x_test.shape[1]))
y_test = to_categorical(y_test)
```

```python
model.save('sentimentmodel.hd5')
```

```python
score = model.evaluate(x_test, y_test, 
                       batch_size=2000, verbose=1)
print('Test score:', score[0])
print('Test accuracy:', score[1])
```
